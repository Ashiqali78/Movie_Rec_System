```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
```

```python
df=pd.read_excel('House_price.xlsx')
df
```

```python
df.isnull().sum()
```

```python
df.Bedroom.median()
```

```python
# Fill the Bedroom column with the median
df['Bedroom'].fillna(df.Bedroom.median(),inplace=True)
```

```python
df['Age'].fillna(df.Age.median(),inplace=True)
```

```python
df
```

```python
lr=linear_model.LinearRegression()
```

```python
# This is training and target variable
lr.fit(df[['Area','Bedroom','Age']],df.Price)
```

```python
lr.predict([[3000,3,20]])
```

```python
lr.predict([[2500,4,5]])
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)  # Create subplots
plt.scatter(df['Area'], df['Price'], marker='o')
plt.plot(df['Area'], lr.predict(df[['Area', 'Bedroom', 'Age']]), color='red')  # Line of best fit
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Area vs Price')

# 2. Bedroom vs. Price
plt.subplot(1, 3, 2)
plt.scatter(df['Bedroom'], df['Price'], marker='o')
plt.plot(df['Bedroom'], lr.predict(df[['Area', 'Bedroom', 'Age']]), color='red')  # Line of best fit
plt.xlabel('Bedroom')
plt.ylabel('Price')
plt.title('Bedroom vs Price')

# 3. Age vs. Price
plt.subplot(1, 3, 3)
plt.scatter(df['Age'], df['Price'], marker='o')
plt.plot(df['Age'], lr.predict(df[['Area', 'Bedroom', 'Age']]), color='red')  # Line of best fit
plt.xlabel('Age')
plt.ylabel('Price')
plt.title('Age vs Price')
```
